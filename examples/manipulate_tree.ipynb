# Manipulating a PALS tree in memory

This notebook reads a PALS/YAML file and then manipulates the resulting tree
structure in memory: inspecting nodes, accessing sequence and map elements,
adding new elements, and writing the edited tree back to a file.

`using PALSJulia` on its own only brings a few names into scope. Calling
`export_manipulators()` additionally exports the tree-manipulation functions
(`parse_file`, `to_yaml_string`, `add_map!`, ...) so they can be called by their
bare names, as done throughout this notebook.

If you would rather **not** pull those extra symbols into your namespace, skip
`export_manipulators()` and instead import the package under an alias:

```julia
import PALSJulia as pj
```

and prefix each call, e.g. `pj.parse_file(ex_file)` instead of `parse_file(ex_file)`.

````julia
using PALSJulia
export_manipulators()

lattice_dir = joinpath(@__DIR__, "..", "lattice_files")
ex_file     = joinpath(lattice_dir, "ex.pals.yaml")
expand_file = joinpath(lattice_dir, "expand.pals.yaml")
````

## Reading a lattice from a YAML file

Use `tree = parse_file(filename)` to read any YAML file. To read a PALS file
*with* lattice expansion, use `parse_and_expand_pals` instead.

````julia
tree = parse_file(ex_file)
````

## Printing a tree

To print a tree to the console, use `to_yaml_string(tree)`.

````julia
println(to_yaml_string(tree))
````

## Type checking

The root node of `ex.pals.yaml` is a sequence, so `is_sequence(tree)` is `true`.

````julia
is_sequence(tree)
````

## Accessing a sequence

Elements in a sequence may be accessed by their index.

````julia
seq1 = tree[1]
println(to_yaml_string(seq1))
````

## Accessing a map

Elements in a map may be accessed by their key.

````julia
map1 = seq1[1]["kind"]
println(to_yaml_string(map1))
````

## Adding new elements

Add a new sequence element to the root containing `new_map: {apples: 5}`.

````julia
new_map_entry = add_map!(tree)
map_node = add_map!(new_map_entry, key="new_map")
add_scalar!(map_node, "5", key="apples")
````

Add another sequence element to the root containing a `magnet_list`:

```yaml
- magnet_list:
    - magnet1
    - magnet2
```

````julia
magnets_entry = add_map!(tree)
sequence = add_sequence!(magnets_entry, key="magnet_list")
add_scalar!(sequence, "magnet1")
add_scalar!(sequence, "magnet2", index=1)
````

## Writing the tree to a file

Use `write_yaml(tree, filename)` to write the edited tree to a file.

````julia
write_yaml(tree, expand_file)
println("Wrote tree to 'expand.pals.yaml'")
````

## The final modified tree

````julia
println(to_yaml_string(tree))
````
